```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
p = print
```

```python
dta=pd.read_csv('bats.csv',skiprows=58, sep="\s+", header=None)
col_names=dta.iloc[0,:].tolist()
dta.columns=col_names
dta.drop(dta.index[0], inplace=True)
#dta=dta.astype(float)
dta['yyyymmdd'] = pd.to_datetime(dta['yyyymmdd'].astype(str), format='%Y%m%d')
dta.iloc[:,4:34]=dta.iloc[:,4:34].astype(float)
dta=dta.replace(-999.0,np.nan)
```
